```python
import os
from langchain.document_loaders import PyPDFLoader, UnstructuredFileLoader

# Step 2: Define document loading function
def load_documents_from_directory(directory_path):
    documents = []
    for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)
        if filename.endswith('.pdf'):
            loader = PyPDFLoader(file_path)
            documents.extend(loader.load())
        elif filename.endswith('.txt'):
            loader = UnstructuredFileLoader(file_path)
            documents.extend(loader.load())
    return documents

# Step 3: Set your actual folder path inside Google Drive
directory_path = "/content/drive/MyDrive/BGSCET"

# Step 4: Load documents
documents = load_documents_from_directory(directory_path)

# Preview the first doc (if loaded)
if documents:
    print(documents[0].page_content[:500])
else:
    print("No documents found in the specified path.")
```

```python
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma

# Chunking
text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = text_splitter.split_documents(documents)

# Embeddings
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Store in ChromaDB
chroma_dir = "/content/drive/MyDrive/BGSCET/chroma_db"
vectordb = Chroma.from_documents(docs, embedding=embedding_model, persist_directory=chroma_dir)
vectordb.persist()

print("✅ Chroma DB created and persisted.")
```

```python
# Assuming you've already loaded and chunked all documents
# Step: Filter chunks that seem like "tales" (based on filename or text content)

tale_docs = [doc for doc in docs if "tale" in doc.page_content.lower() or "folklore" in doc.page_content.lower() or "story" in doc.page_content.lower()]

# Optional: print a preview
print(f"Found {len(tale_docs)} tale-related chunks.")

# Vector DB just for stories
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma

story_embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
story_vectordb = Chroma.from_documents(
    tale_docs, embedding=story_embeddings, persist_directory="chroma_dir"
)
story_vectordb.persist()
```

```python
import pyttsx3
import speech_recognition as sr
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Load model and tokenizer from saved location for fast access
save_path = '/content/drive/MyDrive/BGSCET/story_model'  # Update with your saved model path
bnb_config = "path_to_your_bitsandbytes_config"  # Specify the appropriate bitsandbytes config if needed
model = AutoModelForCausalLM.from_pretrained(save_path, quantization_config=bnb_config, device_map="auto")
tokenizer = AutoTokenizer.from_pretrained(save_path)

# Initialize the text generation pipeline
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Initialize the pyttsx3 TTS engine
engine = pyttsx3.init()

# Function to record audio and convert to text (Speech-to-Text)
def record_audio():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Say something...")
        recognizer.adjust_for_ambient_noise(source)  # Adjust for ambient noise
        audio = recognizer.listen(source)

    try:
        print("Recognizing...")
        # Use the recognize_google method to convert speech to text (can be swapped with any other open-source STT library)
        transcript = recognizer.recognize_google(audio)
        print(f"Recognized Text: {transcript}")
        return transcript
    except sr.UnknownValueError:
        print("Sorry, I couldn't understand that.")
        return None
    except sr.RequestError:
        print("Sorry, there was an issue with the speech service.")
        return None

# Function to generate text from the model
def generate_story(prompt):
    output = pipe(prompt, max_new_tokens=300, do_sample=True, temperature=0.9)
    generated_text = output[0]["generated_text"]
    print(f"Generated Story: {generated_text}")
    return generated_text

# Function to convert text to speech (Text-to-Speech)
def speak_text(text):
    engine.say(text)
    engine.runAndWait()

# Main function
def main():
    # Step 1: Record the user's speech (Speech-to-Text)
    prompt = record_audio()

    if prompt:
        # Step 2: Generate a story using the model
        story = generate_story(prompt)

        # Step 3: Convert the generated story to speech (Text-to-Speech)
        print("Speaking generated story...")
        speak_text(story)

# Run the program
if __name__ == "__main__":
    main()
```
